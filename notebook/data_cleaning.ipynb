### 4. Dropping, Renaming and Reordering Columns

```python
df1 = df1_raw.drop(columns=["Sr No","Remarks","SubVertical","Investors Name"])
df2 = df2_raw.drop(columns=["SubVertical","Investors"])
```

```python
df1 = df1.rename(columns={"Date dd/mm/yyyy":"date", "Startup Name":"startup_name", "Industry Vertical":"industry",
                          "City  Location":"city", "InvestmentnType":"investment_type", "Amount in USD":"amount_usd"})

df2 = df2.rename(columns={"Date":"date", "Startup":"startup_name", "Industry":"industry",
                              "City":"city", "InvestmentType":"investment_type", "InvestmentAmount_USD":"amount_usd"})
```

```python
df2 = df2[["date","startup_name","industry","city","investment_type","amount_usd"]]
```

### 5. Cleaning the Date Column

```python
df1['date'].isnull().sum(), df2['date'].isnull().sum()
```

```python
print(df1['date'].value_counts().to_string())
```

```python
print(df2['date'].value_counts().to_string())
```

```python
df1["date"] = pd.to_datetime(df1["date"],format='mixed',dayfirst=True,errors='coerce')
df2["date"] = pd.to_datetime(df2['date'],format='mixed',dayfirst=True,errors='coerce')
```

```python
df1.isnull().sum(), df2.isnull().sum()
```

```python
df1.dropna(subset=['date'], inplace=True)
```

```python
df1["year"] = df1["date"].dt.year
df2["year"] = df2["date"].dt.year

print("startup_funding.csv — rows per year:")
print(df1["year"].value_counts().sort_index())

print("\n2020_2025_sample.csv — rows per year:")
print(df2["year"].value_counts().sort_index())
```

```python
df1["period"] = df1["year"].apply(lambda x: np.nan if pd.isna(x) else ('Pre-COVID' if x <= 2019 else ('COVID' if x <= 2021 else 'Post-COVID')))
df2["period"] = df2["year"].apply(lambda x: np.nan if pd.isna(x) else ('Pre-COVID' if x <= 2019 else ('COVID' if x <= 2021 else 'Post-COVID')))

df1.drop(columns=["year"], inplace=True)
df2.drop(columns=["year"], inplace=True)

print("\ndf1 period distribution:")
print(df1["period"].value_counts())

print("\ndf2 period distribution:")
print(df2["period"].value_counts())
```

### 7. Cleaning the Industry Column

```python
df1["industry"].isnull().sum(), df2["industry"].isnull().sum()
```

```python
print(df1['industry'].value_counts().to_string())
```

```python
print(df2['industry'].value_counts().to_string())
```

```python
def categorize_industry(value):
    if pd.isna(value) or str(value).strip().lower() == 'nan':
        return np.nan
    
    text = str(value).lower()
    
    if any(keyword in text for keyword in ['fintech', 'finance', 'financial', 'payment', 'lending', 'banking', 'insurance', 'nbfc', 'wallet', 'investment', 'wealth', 'credit', 'loan', 'bfsi', 'p2p']):
        return 'FinTech'
    elif any(keyword in text for keyword in ['edtech', 'education', 'learning', 'elearning', 'skill', 'exam', 'school', 'teaching', 'course', 'tutor']):
        return 'EdTech'
    elif any(keyword in text for keyword in ['health', 'medical', 'healthcare', 'doctor', 'diagnostic', 'pharmacy', 'fitness', 'wellness', 'clinical', 'hospital']):
        return 'HealthTech'
    elif any(keyword in text for keyword in ['food', 'restaurant', 'grocery', 'meal', 'beverage', 'dining', 'kitchen', 'gourmet', 'tiffin', 'catering']):
        return 'FoodTech'
    elif any(keyword in text for keyword in ['logistics', 'delivery', 'shipping', 'freight', 'warehouse', 'supply chain', 'last mile', 'trucking']):
        return 'Logistics'
    elif any(keyword in text for keyword in ['realestate', 'real estate', 'property', 'housing', 'house', 'home', 'accommodation', 'interior']):
        return 'RealEstate'
    elif any(keyword in text for keyword in ['ecommerce', 'e-commerce', 'consumer internet', 'consumerinternet', 'marketplace', 'shopping', 'fashion', 'apparel', 'jewellery', 'jewelry', 'furniture']):
        return 'E-commerce'
    elif any(keyword in text for keyword in ['retail', 'fmcg', 'qsr', 'supermarket', 'store', 'shop', 'offline']):
        return 'Retail'
    elif any(keyword in text for keyword in ['travel', 'hotel', 'rental', 'tourism', 'automobile', 'automotive', 'cab', 'taxi', 'ride', 'transport', 'mobility', 'bus', 'auto rickshaw']):
        return 'Mobility'
    elif any(keyword in text for keyword in ['agri', 'agriculture', 'farm', 'agtech', 'crop']):
        return 'AgriTech'
    elif any(keyword in text for keyword in ['media', 'content', 'publishing', 'news', 'video', 'music', 'entertainment', 'gaming', 'streaming']):
        return 'Media'
    elif any(keyword in text for keyword in ['solar', 'energy', 'cleantech', 'clean tech', 'renewable', 'waste', 'environment']):
        return 'Cleantech'
    elif any(keyword in text for keyword in ['artificial intelligence', 'machine learning', 'deep learning', 'iot', 'internet of things', 'deep-tech', 'deeptech']):
        return 'AI & Deep Tech'
    elif any(keyword in text for keyword in ['electronics', 'devices', 'hardware', 'wearable', 'smartwatch', 'gadget']):
        return 'Consumer Electronics'
    elif any(keyword in text for keyword in ['enterprise', 'b2b', 'crm', 'erp', 'hrtech', 'workforce']):
        return 'Enterprise'
    elif any(keyword in text for keyword in ['saas', 'software', 'cloud', 'platform', 'automation', 'analytics', 'data', 'tech', 'it', 'information technology']):
        return 'SaaS'
    else:
        return 'Others'

df1['industry'] = df1['industry'].apply(categorize_industry)
```
